### Question. Split and train the data 70/30 into training and test datasets.


## Split data set

```python
x=data1_scaled.drop(['Attrition'],axis=1)
y=data1_scaled[['Attrition']]
```

```python
x.head()
```

```python
y.head()
```

```python
print('shape of x is =' , x.shape)
print('shape of y is =' , y.shape)
```

## test train data set

```python
import sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30, random_state=0)
```

```python
print('shape of y_train is =', y_train.shape)
print('shape of y_test is =', y_test.shape)
print('shape of x_train is =', x_train.shape)
print('shape of x_test is =', x_test.shape)
```

### Question:  Build a sequential neural network
3 hidden dense layers
100, 50, 25 nodes
activation function = 'relu'
dropout = 0.5 for each layer

## Now we apply Sequential Neural Network

```python
import tensorflow as tf
from matplotlib.colors import ListedColormap
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Dense(100,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(50,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(25,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
history= model.fit(x_train,y_train.values, validation_data=(x_test, y_test.values),epochs=100)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

### Question: Use early stopping callback to prevent overfitting

## Early Stopping

```python
model = Sequential()

model.add(Dense(100,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(50,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(25,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
callback = EarlyStopping(
    monitor="val_loss",
    min_delta=0.00001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False)
```

```python
history= model.fit(x_train,y_train, validation_data=(x_test, y_test) , epochs=200, callbacks=callback)
```

```python
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

```python
pred= (model.predict(x_test)>0.5).astype('int32')

#pred= (model.predict(x_test))
```

```python
pred
```

## Confusion Mertrics

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
print(cm)
```

#### How to read it

```python
print('\nTrue Positives(TP) = ', cm[0,0])

print('\nTrue Negatives(TN) = ', cm[1,1])

print('\nFalse Positives(FP) = ', cm[0,1])

print('\nFalse Negatives(FN) = ', cm[1,0])
```

```python
from sklearn import metrics
ann_score = metrics.accuracy_score(y_test, pred)
print('accutracy score is =', ann_score * 100,'%' )
```

### Our(SNN) accuracy score is 84.12% percent, which is very good 


```python
plt.figure(figsize=(12,12))
sns.heatmap(cm ,annot=True, fmt='.3f' , linewidths=.5 ,square=True , cmap='Spectral' )
plt.ylabel('Actual value')
plt.xlabel('Predecited value')
all_sample_title ='ANN model accuracy : {0}% '.format( ann_score * 100)
plt.title(all_sample_title, size  =17)
```

## ROC and AUC Value

```python
auc = metrics.roc_auc_score(y_test, pred)

print('ROC and AUC Value =',auc)
```

ROC AUC is a single number summary of classifier performance. The higher the value, the better the classifier.

ROC AUC of our model approaches towards 1.But here it gives 0.5

## ROC Graph

```python
fpr, tpr, _ = metrics.roc_curve(y_test,  pred)

#create ROC curve
plt.plot(fpr,tpr)
plt.plot([1,0], [1,0], 'k--' )
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

## AUC Graph

```python
fpr, tpr, _ = metrics.roc_curve(y_test,  pred)
auc = metrics.roc_auc_score(y_test, pred)

#create ROC curve
plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.plot([1,0], [1,0], 'k--' )
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.show()

```
